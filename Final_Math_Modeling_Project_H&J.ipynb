# **Graph 1: First Round of Data**


*The graph below shows our first data collecting session, the number of reps versus the weight of the dumbbell.*

```
#define data
data = StringIO("""x	y
10	40
15	38
20	36
25	34
30	27
35	28
40	20
45	28
50	22
60	19
70	14
""")

# convert data to dataframe
df = pd.read_csv(data, sep ="\t")

# define arrays for the data
xdata = df['x']
ydata = df['y']

# create graph
fig1 = plt.figure(figsize=(12,8))

# define the model function; this function returns y(x) for a given x and fit parameters
model = lambda x,m,b: m*x+b

# Find best fit parameters
guess=[1,1] # this is the initial guess for the fit parameters
params, covariance = curve_fit(model, xdata, ydata, p0=guess) #curve fit takes our mode, data, and guess

# calculate points for the model
xmodel = np.linspace(0,110)
ymodel = model(xmodel,*params)


# plot data and model
plt.plot(xdata,ydata,'bo-') # data
plt.plot(xmodel,ymodel,'y',label='Automatic Line of Best Fit')
plt.grid()
plt.title("Reps vs Weight - First Round of Data")
plt.xlabel("Weight (LBS)")
plt.ylabel("Reps")
plt.legend()
plt.show()

print("Pounds: 10, 15, 20, 25, 30, 35, 40 = Monday, Nov 16 @ 4pm, 5pm, 6pm, 7pm, 8pm, 9pm, 10pm")
print("Pounds: 45, 50, 60, 70 = Tuesday, Nov 17 @ 5pm, 6pm, 7pm, 8pm")

print("The curve fit parameters are: ", *params)
```

The equation for the first round of data for reps vs weight:

$$y = -0.43x + 43.34$$

#**Graph 6: Training Volume**
The next graph shows training volume. Training volume is calculated by taking the number of sets multiplied by reps multiplied by weight. For our data collecting there was one set. Therefore, we used reps multiplied by weight to calculate our training volume.

In a study conducted in Norway they analyzed 27 Norwegian Rowers who were winning international medals in 1970-2001. They analyzed their workouts through out the 31 years and compared them to how many medals they were winning. They found that their rowing performance increased on average over 10% during the 31 years and they found that the training volume of their workouts typically increased by 20%. This study analyzed many possible different factors, training volume just being one, but they concluded that the data was consistent with what has been discovered in other sports, that with workouts of higher training volume, their rowing improved. This study demonstrates that training volume can definitely have an impact on strength and training.


http://bionics.seas.ucla.edu/education/Rowing/Training_2004_01.pdf


```
#Create a graph with average of both data
#first round of data
data = StringIO("""x	y
10	415
15	585
20	770
25	900
30	900
35	1032.5
40	900
45	1192.5
50	1075
60	1110
70	1015
""")

# convert data to dataframe
df = pd.read_csv(data, sep ="\t")

# define arrays for the data
xadata = df['x']
ybdata = df['y']

# create graph
fig1 = plt.figure(figsize=(12,8))

# define the model function; this function returns y(x) for a given x and fit parameters
model = lambda x,a,b: a*x**2+b*x

# Find best fit parameters
guess=[1,1] # this is the initial guess for the fit parameters; you need a 1 for each parameter
params, covariance = curve_fit(model, xadata, ybdata, p0=guess)

# calculate points for the model
xamodel = np.linspace(0,70,1000)
ybmodel = model(xamodel,*params)

b=params[1]
a=params[0]
xpeak=-b/(2*a)

# plot data and model
plt.plot(xadata,ybdata,'bo-', label='Average of both Data') # data
plt.plot(xamodel,ybmodel,'r', label='Automatic Line of Best Fit')
plt.grid()
plt.title("Training Volume - Average of First & Second")
plt.xlabel("Weight (LBS)")
plt.ylabel("Training Volume")
plt.ylim(0,1200)
plt.legend()
plt.show()

print("We used the quadratic formula to create the automatic line of best fit.")
print("The curve fit parameters are: ", *params)
print("The max training volume = {:.2f} pounds at a lifting weight of {:.2f} pounds.".format(np.max(ybmodel), xpeak))
```

We believe that the training volume and weight at the peak would show which weight and how many reps would have the most and best effect in your workout.
